```python
import regex as re
import json
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
os.chdir(r'C:\Users\jschu\GitHub\Portfolio\Personal_Cocktail\Diffords')

with open('cocktail_detail.json') as f:
    detail = json.load(f)
    
with open('cocktail_classes.json') as f:
    types = json.load(f)
```

```python
#Clean up types dictionary - there's duplicates in the dataset
#Create new dictionary called cocktail_class_types. This will include only 
cocktail_class_types = {}

for class_name in types.keys():
    find_dup = []
    for name in types[class_name]['name']:
        if name not in find_dup:
            find_dup.append(name)
    cocktail_class_types[class_name] = find_dup
```

```python
#Get detail json into DataFrame
detail_df = pd.read_json('cocktail_detail.json')
detail_df = detail_df.transpose()
```

```python
#drop some unneccessary columns
detail_df.drop('image',axis = 1,inplace = True)
detail_df.drop('url',axis = 1,inplace = True)
```

```python
detail_df.info()
```

```python
check_hist_booz = detail_df['booziness'].dropna()
check_hist_dry = detail_df['dryness'].dropna()
#plt.hist(check_hist_booz)
#plt.hist(check_hist_dry)
sns.displot(check_hist_booz,bins = 10)
sns.displot(check_hist_dry, bins = 10)
```

```python
#Fill data without booziness as mean value
#Fill data without dryness as mean value
detail_df['booziness'].fillna(detail_df['booziness'].mean(),inplace=True)
detail_df['dryness'].fillna(detail_df['dryness'].mean(),inplace=True)
detail_df.head()
```

```python
#Split up ingredients into quantities and 
ingredient_quant = [i.values() for i in detail_df['ingredient']]
ingredient_name = [i.keys() for i in detail_df['ingredient']]
detail_df['ingredient_quant'] = ingredient_quant
detail_df['ingredient_name'] = ingredient_name
detail_df.drop('ingredient',axis = 1,inplace = True)
detail_df.info()
```

```python
#Reorganize Class variable. Some names appear more than once in each class.

cocktail_name = {}
for name in detail_df.index:
    cocktail_name[name] = []

unique_list = []
not_in_list = []
    
for count,ctype in enumerate(cocktail_class_types.keys()):
    for name in cocktail_class_types[ctype]:
        if name not in unique_list:
            unique_list.append(name)
        try:
            if type(cocktail_name[name]) == list:
                cocktail_name[name].append(count)
            #elif cocktail_name[name] == None:
            #    cocktail_name[name] = [ctype]
        except:
            if name not in not_in_list:
                not_in_list.append(name)
                
print('Not in List: ' + str(len(not_in_list)))
print('Total DB len: ' + str(len(cocktail_name)))
print('Unique parsed: ' + str(len(unique_list)))

#print(not_in_list)
```

```python
#Add the class details into the detail_df database. Replace the original variable class
detail_df['class'] = [values for values in cocktail_name.values()]

#Attribute various classes to each cocktail. This will be potentially used as our label vector
count = 0
for index in detail_df.index:
    if cocktail_name[index] != detail_df['class'].loc[index]:
        count+=1
        print(index)
print(count)
```

```python
#What are the cocktail class types?
#print(cocktail_class_types.keys())
print('Length of different types: ' + str(len(cocktail_class_types.values())))
class_len_list = [len(cocktail_class_types[types]) for types in cocktail_class_types]
print(class_len_list)

fig,ax = plt.subplots(1,1,figsize = (20,5))
ax.bar(cocktail_class_types.keys(), height = class_len_list)
plt.xticks(rotation=90)
```
