```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = r'C:\Users\jschu\OneDrive\Documents\MIDS\W200 Intro to Programming\Project 2\heart.csv'
heart = pd.read_csv(file)
```

```python
heart.ExerciseAngina.replace({'N':0,'Y':1},inplace = True)
heart.Cholesterol.replace({0:None},inplace = True)
heart.ChestPainType.replace({'TA':1,'ATA':2,'NAP':3,'ASY':4},inplace=True)
heart.Sex.replace({'M':1,'F':0},inplace=True)
heart.RestingECG.replace({'Normal':0,'ST':1,'LVH':2},inplace=True)
heart.ST_Slope.replace({'Up':1,'Flat':2,'Down':3},inplace=True)
heart.RestingBP.replace({0:None},inplace = True)
```

```python
#Fill missing cholestrol value by the mean value

heart.Cholesterol.fillna(heart.Cholesterol.mean(),inplace=True)
heart.RestingBP.fillna(heart.RestingBP.mean(),inplace = True)
heart.isnull().sum()
```

```python
for i in ['Age','RestingBP','FastingBS','MaxHR']:
    nheart = list(heart[heart.HeartDisease == 0][i].dropna())
    yheart = list(heart[heart.HeartDisease == 1][i].dropna())
    xmin = min(min(nheart),min(yheart))
    xmax = max(max(nheart),max(yheart))
    width = (xmax-xmin)/40
    sns.distplot(nheart,color = 'r', kde = False, bins = np.arange(xmin,xmax,width))
    sns.distplot(yheart,color = 'g', kde = False, bins = np.arange(xmin,xmax,width))
    plt.legend(['No Heart Disease','Yes Heart Disease'])
    plt.title('Overlaid histogram for {}'.format(i))
    plt.show()
```
